# UPLOAD DAN IMPORT DATA

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split #Untuk percentage split
from sklearn.model_selection import cross_val_score, cross_val_predict #k-fold cross validation
from sklearn.svm import SVC #library algoritma svm menggunakan kernel
```

```python
data = pd.read_csv('data.csv') #panggil dataset
data.head()
```

# Mengubah Label Object menjadi Kategori


```python
data.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
df = pd.DataFrame(data)
# Membuat objek LabelEncoder
label_encoder = LabelEncoder()
df['diagnosis'] = label_encoder.fit_transform(df['diagnosis'])
df
```

# Pre-Processing

```python
df = df.drop(columns=['id', 'Unnamed: 32'])
df
```

```python
import seaborn as sea
sea.set()
#sea.pairplot(df, hue='diagnosis', height=1)
```

```python
X_lama= df.drop(columns='diagnosis', axis=1)
Y_lama= df['diagnosis']
```

```python
print(df.groupby('diagnosis').size())
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=0)
X, Y = ros.fit_resample(X_lama, Y_lama)
print("Shape of X_resampled:", X.shape)
print("Shape of y_resampled:", Y.shape)
print("Class distribution of y_resampled:")
print(pd.Series(Y).value_counts())
```

# NORMALISASI



```python
X.describe ()
```

```python
df_norm=(X-X.min())/(X.max()-X.min())
df_norm
```

```python
Y.info()
```

# DATA TESTING

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# Aktivasi Fungsi SVM, Menggunakan Kernel Linear
# Menggunakan Metode Training SPLIT

```python
svm = SVC(kernel='linear')
svm
```

```python
svm.fit(X_train, y_train)
```

```python
y_predictsvm = svm.predict(X_test)
y_predictsvm
```

```python
print('Confusion Matrix: ', (confusion_matrix(y_test, y_predictsvm))) #Membuat confusion matrix
print('Classification Report', (classification_report(y_test, y_predictsvm)))#Membuat classification report
print('Akurasi: ', (accuracy_score(y_test, y_predictsvm)))#Menghitung akurasi model SVM
```

```python
# SIMPAN MODEL
import pickle
filename= 'breastcancer_model.sav'
pickle.dump (y_predictsvm, open(filename, 'wb'))
```

# Menggunakan Metode Training K-VOLD

```python
svm2 = SVC(kernel='linear')
svmscore = cross_val_score(svm2, X, Y, scoring='accuracy', cv=100)
y_predictsvm2 = cross_val_predict(svm2, X, Y, cv=100)
y_predictsvm2
```

```python
print('Confusion Matrix:', (confusion_matrix(Y, y_predictsvm2))) #Membuat confusion matrix
print('Classification Report: ', (classification_report(Y, y_predictsvm2))) #Menghitung akurasi model SVM
print('Akurasi: ', (accuracy_score(Y, y_predictsvm2)))#Membuat classification report
```
